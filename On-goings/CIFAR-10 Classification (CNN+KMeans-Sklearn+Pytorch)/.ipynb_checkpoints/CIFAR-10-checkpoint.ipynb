```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data as data
from torch.autograd import  Variable
import torchvision
from torchvision import models, transforms
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pickle
import os
```

```python
channels = 3
IMG_SIZE = 32
device = torch.device("cuda: 0" if torch.cuda.is_available() else "cpu")
```

```python
DATA_PATH = { "train": "data/train", 
              "test": "data/test",
             "label_names" : "data/batches.meta"}
```

```python
# Load label names
with open(DATA_PATH["label_names"], 'rb') as fo:
    c = pickle.load(fo, encoding='bytes')
CLASSES = [str(i,'utf-8') for i in c[b"label_names"]]
NUM_CLASSES = len(CLASSES)
```

```python
# Load CIFAR dataset
def unpickle(str_type):
    print("Loading data...")
    filename = os.listdir(DATA_PATH[str_type])
    imgs = []
    labels = []
    
    for i in filename:
        path = os.path.join(DATA_PATH[str_type],i)
        with open(path, 'rb') as fo:
            d = pickle.load(fo, encoding='bytes')
            imgs.append(d[b"data"])
            labels += d[b"labels"]
    print("Data loaded!")
    return np.array(imgs).reshape(-1,channels,IMG_SIZE,IMG_SIZE), np.array(labels)

imgs_train, labels_train = unpickle("train")
imgs_val, labels_val = unpickle("test")
```

```python
print(f"Number of training samples: {len(labels_train)}")
print(f"Number of vaidating samples: {len(labels_val)}")
print(f"Number of unique labels : {NUM_CLASSES}")
print(f"Unique labels name: \n{CLASSES}")
```

```python
# Visualize some training samples
from mpl_toolkits.axes_grid1 import ImageGrid

def visualize_samples(NUM_PIC=10):
    a = np.random.randint(100-NUM_PIC)
    fig = plt.figure(figsize=(NUM_PIC*2,NUM_PIC*2))
    plt.title("Training Samples", fontsize=25)
    plt.xticks([])
    plt.yticks(np.arange(10)[-1:-11:-1]+0.3, CLASSES, fontsize=25)
    grid = ImageGrid(fig, 111, (NUM_PIC,NUM_PIC), axes_pad=0)
    
    imgs_id_list = []
    for i in range(10):
        imgs_id = np.where(labels_train == i)
        imgs_id_list += list(imgs_id[0][a:a+NUM_PIC])
    
    for ax, im_id in zip(grid, imgs_id_list):
        img = imgs_train[im_id]
        ax.imshow(img.transpose((1,2,0)))
        ax.axis("off")
    
    
    plt.show()
    
    
visualize_samples()
```

```python
def visualize_frequency():
    id_count = {CLASSES[i]: len(np.where(labels_train == i)[0]) for i in range(10)}

    fig = plt.subplots(figsize=(15,10))
    plt.bar(id_count.keys(), height=id_count.values())
    plt.title("Frequency of each class", fontsize= 20)
    plt.xlabel("Class", fontsize= 20)
    plt.ylabel("Number of samples", fontsize= 20)
    plt.show()
    
visualize_frequency()    
```

```python
# Custom CNN model

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv_layer = nn.Sequential(

            # Conv Layer block 1
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),

            # Conv Layer block 2
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout2d(p=0.05),

            # Conv Layer block 3
            nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )


        self.fc_layer = nn.Sequential(
            nn.Dropout(p=0.1),
            nn.Linear(4096, 1024),
            nn.ReLU(inplace=True),
            nn.Linear(1024, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.1),
            nn.Linear(512, 10)
        )


    def forward(self, x):
        """Perform forward."""
        
        # conv layers
        x = self.conv_layer(x)
        
        # flatten
        x = x.view(x.size(0), -1)
        
        # fc layer
        x = self.fc_layer(x)

        return x
```

```python
class CifarDataset(data.Dataset):
    def __init__(self, tensor, labels, transforms = None, device = None):
        self.tensor = tensor
        self.labels = labels
        self.transforms = transforms
        self.device = device
        if self.transforms is not None:
            for ids in range(len(self.tensor)):
                img = Image.fromarray(self.tensor[ids].reshape(IMG_SIZE,IMG_SIZE,3))
                self.tensor[ids] = self.transforms(img)
        
        if self.device is not None:
            self.tensor = torch.Tensor(self.tensor).to(device)
            self.labels = torch.LongTensor(self.labels).to(device)
            
    def __getitem__(self, ids):
        return self.tensor[ids], self.labels[ids]
    
    def __len__(self):
        return self.tensor.shape[0]
```

```python
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(IMG_SIZE, padding = 4),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
])

#img_train_tensor = torch.Tensor(imgs_train)
#label_train_tensor = torch.LongTensor(labels_train)

#img_val_tensor = torch.Tensor(imgs_val)
#label_val_tensor = torch.LongTensor(labels_val)
```

```python
trainset = CifarDataset(imgs_train,labels_train, train_transform, device)
valset = CifarDataset(imgs_val,labels_val, val_transform, device)
```

```python
BATCH_SIZE = 256
trainloader = data.DataLoader(trainset, batch_size=BATCH_SIZE, num_workers=0)
valloader = data.DataLoader(valset, batch_size=BATCH_SIZE, num_workers=0)
```

```python
def correct_count(output, labels):
    predictions = torch.max(output, 1)[1]
    correct = (predictions==labels).sum()
    return correct
```

```python
def train_step(model, optimizer, criterion, trainloader):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    for i, (img,label) in enumerate(trainloader):
        optimizer.zero_grad()
        output = model(img).squeeze()
        label = label.squeeze()
        loss = criterion(output,label)
        
        loss.backward()
        optimizer.step()
        total_loss += loss.data
        correct += correct_count(output, label)
        total += len(label)
    accuracy = correct*1.0/total
    return model, total_loss, accuracy
```

```python
def evaluate_step(model, criterion, valloader):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for i, (img,label) in enumerate(valloader):
            output = model(img).squeeze()
            label = label.squeeze()
            loss = criterion(output,label)

            total_loss += loss.data
            correct += correct_count(output, label)
            total += len(label)
        accuracy = correct*1.0/total
    return total_loss, accuracy
```

```python
TRAIN_DATA = 1

loss_list = {
    "train": [],
    "val": []
}

acc_list = {
    "train": [],
    "val": []
}


if TRAIN_DATA:
    EPOCHS = 50
    print("Start training...")
    for epoch in range(EPOCHS):
        model, train_loss, train_acc = train_step(model, optimizer, criterion,trainloader)
        val_loss, val_acc = evaluate_step(model, criterion,valloader)
        loss_list["train"].append(train_loss)
        loss_list["val"].append(val_loss)
        acc_list["train"].append(train_acc)
        acc_list["val"].append(val_acc)
        print(f"Epoch: [{epoch+1}|{EPOCHS}] \n\t Training Loss: {train_loss:.3f} | Training Acc: {train_acc:.3f}\n\t Validating Loss: {val_loss:.3f} | Validating Acc: {val_acc:.3f}")
        torch.save(model.state_dict(), "model/modelCNN.pt")
    print("Training Completed!")
```
