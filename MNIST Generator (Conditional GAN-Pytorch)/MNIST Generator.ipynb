```python
import torch
import torchvision
from torch.autograd import Variable
import torch.nn as nn
import torch.utils.data as data
from torchvision import datasets, transforms, models
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load data
dataset = datasets.MNIST("data",train=True, download= True, transform=data_transform)
dataloader = data.DataLoader(dataset, batch_size=BATCH_SIZE, num_workers=4, shuffle=True)
```

```python
LATENT_SIZE = 100
```

```python
#Generate labels for loss fuction
def generate_labels(size, label):
    if label:
        data = torch.ones(size) 
    else:
        data = torch.zeros(size)
    data = Variable(data).to(device)
    return data.view(-1)
```

```python
one_hot = np.eye(10)
```

```python
def generate_fake_classes(size):
    data = torch.LongTensor(np.random.randint(0,10,size))
    data = Variable(data).to(device)
    return data
```

```python
class Generator(nn.Module):
    def __init__(self, inp, out):
        super(Generator, self).__init__()
        self.label_emb = nn.Embedding(10, 10) #Can be increased for improvement
        self.main1 = nn.Sequential(
            nn.Linear(inp+10,256),
            nn.LeakyReLU(0.2, inplace= True),
   
            nn.Linear(256, 512),
            nn.LeakyReLU(0.2, inplace= True),
            
            nn.Linear(512, 1024),
            nn.LeakyReLU(0.2, inplace= True),
            
            nn.Linear(1024, out),
            nn.Tanh()
        )
    
    def forward(self, x ,y):
        y = self.label_emb(y)
        x = torch.cat([x, y], 1)
        x = self.main1(x)
        return x.view(-1,28,28)
```

```python
class Discriminator(nn.Module):
    def __init__(self, inp, out):
        super(Discriminator, self).__init__()
        
        self.label_emb = nn.Embedding(10, 10) #Can be increased for improvement
        
        self.main1 = nn.Sequential(
            nn.Linear(inp+10, 1024),
            nn.LeakyReLU(0.2, inplace= True),
   
            nn.Linear(1024,512),
            nn.LeakyReLU(0.2,inplace = True),
            nn.Dropout(0.3),
            
            nn.Linear(512,256),
            nn.LeakyReLU(0.2,inplace = True),
            nn.Dropout(0.3),
            
            nn.Linear(256,1),
            nn.Sigmoid()
        )
        
    def forward(self, x, y):
        y = self.label_emb(y)
        x = x.view(-1, IMG_SIZE*IMG_SIZE)
        x = torch.cat([x, y], 1)
        x = self.main1(x)    
       
        return x.view(-1)
    
```

```python
#Make latent vector by creating random Gaussian noise
def random_noise(size):
    noise = torch.randn(size,LATENT_SIZE)
    noise = Variable(noise).to(device)
    return noise
```

```python
def train_discriminator(D, G, optimizer, error, real_batch, real_classes_batch):
    batch_size = real_batch.size(0)
    optimizer.zero_grad()
    
    real_classes = Variable(real_classes_batch.to(device))
    real_data = Variable(real_batch.view(-1,IMG_SIZE*IMG_SIZE)).to(device)
    prediction_real = D(real_data,real_classes)
    real_label = generate_labels(batch_size,1)
    loss_real = error(prediction_real,real_label)
    
    fake_classes = generate_fake_classes(batch_size)
    fake_data = G(random_noise(batch_size), fake_classes).detach().to(device)
    
    fake_classes = generate_fake_classes(batch_size)
    prediction_fake = D(fake_data,fake_classes)
    fake_label = generate_labels(batch_size,0)
    loss_fake = error(prediction_fake, fake_label)
    
    loss = loss_fake + loss_real
    loss.backward()
    
    D_x = prediction_real.mean().item()
    D_g_z1 = prediction_fake.mean().item()
    optimizer.step()

    return loss_real.data + loss_fake.data, D_x, D_g_z1
```

```python
def train_generator(D,G,  optimizer, error, real_batch):
    batch_size = real_batch.size(0)
    optimizer.zero_grad()
    
    fake_classes = generate_fake_classes(batch_size)
    fake_data = G(random_noise(batch_size),fake_classes).to(device)
 
    prediction = D(fake_data, fake_classes)
    label = generate_labels(batch_size, 1)
    loss = error(prediction, label)
    loss.backward()
    D_g_z2 = prediction.mean().item()
    optimizer.step()
    return loss.data, D_g_z2
```

```python
def generate_img(G, noise):
    G.eval() 
    fig = plt.figure(figsize=(10,3))
    plt.title("Generated Images")
    plt.axis("off")
    plt.tight_layout()
    with torch.no_grad():
        labels = torch.LongTensor([i for i in range(10)]).to(device)
        img = G(noise,labels).cpu().numpy()
        for id, i in enumerate(img):
            fig.add_subplot(2,5,id+1)
            plt.axis("off")
            plt.imshow(i,cmap="gray")
    plt.show()
    G.train()
```

```python
#Generate latent vector
test_noise = random_noise(10)
```

```python
def train(D, G, D_optimizer, G_optimizer, error, trainloader):
    EPOCHS = 100
    D_loss_list = []
    G_loss_list = []
    for epoch in range(EPOCHS):
        for i, (real_batch,real_classes_batch) in enumerate(trainloader):  
            D_loss,D_x, D_g_z1 = train_discriminator(D, G, D_optimizer,error, real_batch, real_classes_batch)
            G_loss, D_g_z2 = train_generator(D,G, G_optimizer,error, real_batch)
            if i%300 == 0 :
                print("Epoch: ({}/{}), Batch: ({}/{}), D_Loss: {:.4f}, G_Loss: {:.4f}, D(x): {:.4f}, G(D(z)): {:.4f} / {:.4f}".format(epoch+1,EPOCHS,i,len(trainloader),D_loss,G_loss,D_x, D_g_z1,D_g_z2))
            D_loss_list.append(D_loss)
            G_loss_list.append(G_loss)   
        if epoch%5==0:
            torch.save(G.state_dict(), "model/generator-cgan2.pth")
            torch.save(D.state_dict(), "model/discriminator-cgan2.pth")
            generate_img(G,test_noise)
    #plot(D_loss_list,G_loss_list,len(D_loss_list))
    return G, D_loss_list,G_loss_list
```

```python
G = Generator(100,784).to(device)
D = Discriminator(784,1).to(device)
if TRAIN_DATA:
    D_optimizer = torch.optim.Adam(D.parameters(), lr= 1e-4)
    G_optimizer = torch.optim.Adam(G.parameters(), lr = 1e-4)
    error = nn.BCELoss()
    print("Start Training.....")
    G, D_loss_list, G_loss_list = train(D,G,D_optimizer,G_optimizer,error,dataloader)
    print("Training Completed!.....")
```

```python
def plot(D_loss_list, G_loss_list, EPOCHS=None):
    if EPOCHS is None:
        EPOCHS = len(D_loss_list)
    fig = plt.figure(figsize=(15,5))
    plt.plot(range(EPOCHS), D_loss_list, label = "Discriminator Loss")
    plt.plot(range(EPOCHS),G_loss_list, color = "orange", label = "Gennerator Loss")
    plt.legend()
    plt.show()
    
plot(D_loss_list, G_loss_list)
```
